```python
import pandas as pd
import FBCrawling
import datetime
import time
```

```python
with open('acc.txt','r') as f:
    inf = f.read()
    
user,pw = inf.split('\n')
    
dp="./driver/chromedriver.exe"
```

```python
loger = login.log_in(driver_path=dp,user=user,pw=pw)
```

```python
driver = loger.open_fb()
```

```python
time_scrapt = datetime.datetime.now().strftime('%d/%m/%Y %H:%M:%S')
master_res = pd.DataFrame()

looping_time = 1e3
# looping_time = 10


count_cnt = 0

for i in range(int(looping_time)):
#     driver.refresh()
    time.sleep(2)
    crawler = login.crawl_info(driver=driver)
    
    try:
        all_cnt = crawler.crawl_cnt_2(0)[3]
        if count_cnt < len(all_cnt):
            try:
            #     Crawl master overall content:
                cnt = crawler.crawl_cnt_2(count_cnt)
                try:
                    see_more = crawler.see_more_button(cnt[1])
                    see_more.click()
                    cnt = crawler.crawl_cnt_2(count_cnt)
                except: pass

            #     Writer name:
                wrt = crawler.crawl_name(cnt[1])

            #     Time passed:
                time_pas = crawler.crawl_time(cnt[1])

            #     Create result:
                res = crawler.crawl_feeling(cnt[1])
                res.update({"writer":wrt,"content":cnt[0],"time_pass":time_pas,"time_crawl":time_scrapt})
                df = pd.DataFrame(res,index=[0])

            #     Append to master result:
                if len(master_res) != 0:
                    if len(master_res[(master_res['writer'] == df['writer'].values[0]) &
                               (master_res['content'] == df['content'].values[0])]) != 0:

                        pass
                #                 (master_res['time_crawl'] = df['time_crawl'].values[0])
                    else:
                        master_res = master_res.append(df,ignore_index=True)
                else:
                    master_res = master_res.append(df,ignore_index=True)
            except: pass
            count_cnt+=1
        else:
            crawler.scrolling()
            count_cnt+=1
    except:
        driver.refresh()
```
